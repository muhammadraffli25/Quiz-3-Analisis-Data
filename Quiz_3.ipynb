# Quiz 3.1: Analisis Deskriptif Data Angka Harapan Hidup (AHH)
Data ini berisi statistik Angka Harapan Hidup (AHH) per wilayah. Dalam notebook ini, kita akan melakukan *import* data, melihat *preview* data, melakukan analisis statistik deskriptif dasar (seperti rata-rata, nilai minimum, maksimum), dan melihat sebaran (distribusi) datanya menggunakan visualisasi grafik (Histogram).

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 3. Import Data ---
# Kita pakai data AHH yang tadi (ingat pemisahnya pakai sep='\t' atau hapus sep-nya kalau sudah diganti koma)
df = pd.read_csv('AHH.csv', sep='\t')

# --- 4. Preview Datanya ---
print("--- PREVIEW DATA ---")
display(df.head())

# --- 5. Lakukan Analisis Deskriptif ---
# Fungsi describe() akan otomatis menghitung rata-rata, min, max, kuartil dari data angkamu
print("\n--- ANALISIS DESKRIPTIF ---")
display(df.describe())

# --- 6. Cek Distribusi Datanya ---
# Kita buat grafik Histogram untuk melihat sebaran nilai AHH-nya
print("\n--- GRAFIK DISTRIBUSI DATA AHH ---")
plt.figure(figsize=(8, 5))

# Mengambil kolom angka (kolom ke-2 / index 1) untuk dibuatkan grafik
# Ubah warna grafik sesukamu, di sini aku pakaikan warna biru langit (skyblue)
plt.hist(df.iloc[:, 1], bins=10, color='skyblue', edgecolor='black')

plt.title('Distribusi Data Angka Harapan Hidup', fontsize=14)
plt.xlabel('Nilai Angka Harapan Hidup', fontsize=12)
plt.ylabel('Frekuensi (Jumlah Wilayah)', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
